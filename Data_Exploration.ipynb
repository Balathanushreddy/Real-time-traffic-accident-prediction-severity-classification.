Identifying the Features

```python
numeric_features = df.select_dtypes(include=['int64', 'float64']).columns
categorical_features = df.select_dtypes(include=['object']).columns
datetime_features = df.select_dtypes(include=['datetime64']).columns
text_features = df.select_dtypes(include=['object']).columns
boolean_features = df.select_dtypes(include=['bool']).columns
```

```python
print("Numeric:", numeric_features.tolist())
print("Categorical:", categorical_features.tolist())
print("Datetime:", datetime_features.tolist())
print("Boolean:", boolean_features.tolist())
print("Text:",text_features.tolist())
```
