we are going to clean data and preprocess the data to create new columns which will help us to make better predictions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('/content/Reported_Crash_Locations.csv')
```

```python
train.info()
```

```python
train.drop('X',axis=1,inplace=True)
```

```python
train.info()
```

```python
train.head()
```

```python
train.drop('Y',axis=1,inplace=True)
```

```python
train.set_index('OBJECTID')
```

We are going to separate dateofcrash into separate columns

```python
train['DateOfCrash'] = pd.to_datetime(train['DateOfCrash'],utc=True)
```

```python
train['Date'] = train['DateOfCrash'].dt.date
train['Time'] = train['DateOfCrash'].dt.time
```

```python
train.drop('DateOfCrash',axis=1,inplace=True)
```

```python
train.head()
```

```python
train.info()
```

```python
train.drop('key_crash',axis=1,inplace=True)
```

```python
train.drop('LocalUse',axis=1,inplace=True)
```

```python
train.drop('UpdateDate',axis=1,inplace=True)
```

```python
train.head()
```

We are going to LocationFeetFromRoad and LocationMilesFromRoad into one clean column called DistanceFromRoad

```python
train['FeetToMiles'] = train['LocationFeetFromRoad'] / 5280
train['DistanceFromRoad'] = np.where(
    train['LocationMilesFromRoad'].notna(),
    train['LocationMilesFromRoad'],
    train['FeetToMiles'])
train.drop(['FeetToMiles', 'LocationFeetFromRoad', 'LocationMilesFromRoad'], axis=1, inplace=True)
```

```python
train.info()
```

we are going to make one column 'roadRelation'

```python
def simplify_road_relation(row):
  if str(row['LocationRampIndicator']).strip().upper() in ['Y','Yes','True']:
    return 'On/Off ramp'
  if str(row['LocationAtFromIndicator']).strip().lower() in ['at','from','to']:
    return 'At Intersection'
  if str(row['LocationInNearIndicator']).strip().upper() in ['IN','NEAR']:
    return 'Near Road'
  rel = str(row['LocationRelationToRoad']).strip().lower()
  if 'on' in rel:
    return 'On Road'
  elif 'off' in rel:
    return 'Off Road'
  return 'Unknown'
train['RoadRelation'] = train.apply(simplify_road_relation,axis=1)
```

```python
train.drop([
    'LocationRelationToRoad',
    'LocationInNearIndicator',
    'LocationRampIndicator',
    'LocationAtFromIndicator'
], axis=1, inplace=True)
```

```python
train.info()
```

```python
train.drop([
    'LocationRoadNameTo',
    'LocationRoadNameAt',
    'LocationRoadNameOn',
], axis=1, inplace=True)
```

```python
train.info()
```

```python
train.drop([
    'WeatherCondition2',
    'unknown_person_type',
], axis=1, inplace=True)
```

```python
train.info()
```

```python
train.head()
```

```python
train.to_csv("cleaned_crash_data.csv", index=False)
```
